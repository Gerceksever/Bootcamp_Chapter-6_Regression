```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from scipy import stats
import statsmodels.api as sm
from matplotlib.ticker import FuncFormatter
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
sicaklik = pd.read_csv("weatherHistory.csv")
```

```python
sicaklik['Fark'] = sicaklik['Apparent Temperature (C)'] - sicaklik['Temperature (C)']
X = sicaklik[['Humidity', 'Wind Speed (km/h)']]
y = sicaklik['Fark']
```

```python
lrm = LinearRegression()
lrm.fit(X, y)
print('Değişkenler: \n', lrm.coef_)
print('Sabit değer (bias): \n', lrm.intercept_)
```

```python
X = sm.add_constant(X)
results = sm.OLS(y, X).fit()
results.summary()
```

p değerleri 0 olduğu için istatistiksel olarak anlamlıdır. Nem ve Rüzgar hızı ile orantılı artış sözkonusu

```python
sicaklik['Fark'] = sicaklik['Apparent Temperature (C)'] - sicaklik['Temperature (C)']
sicaklik['nem_ruzgar'] = sicaklik['Humidity'] * sicaklik['Wind Speed (km/h)']
X = sicaklik[['Humidity', 'Wind Speed (km/h)', 'nem_ruzgar']]
y = sicaklik['Fark']

X = sm.add_constant(X)

sonuclar = sm.OLS(y, X).fit()
sonuclar.summary()
```

```python
# the first look through the data 

Train = pd.read_csv("train.csv", encoding = "ISO-8859-1")
Train.head()
```

```python
np.abs(Train.corr().iloc[-1, 0:-1]).sort_values(ascending=False)
```

```python
def show_missing (df):
    """This function returns percentage and total number of missing values"""
    percent = df.isnull().sum()*100/df.shape[0]
    total = df.isnull().sum()
    missing = pd.concat([percent, total], axis=1, keys=['percent', 'total'])
    return missing[missing.total>0].sort_values('total', ascending=False)
```

```python
Missing=show_missing(Train)
```

```python
Traindrop=Train.dropna(subset=list(Missing[Missing.total<10].index))
```

```python
Traindrop.drop(list(Missing[Missing.percent>20].index), axis=1, inplace=True)
```

```python
Corr_List=list(np.abs(Train.corr().iloc[-1,0:-1]).sort_values(ascending=False).index[0:20])
```

```python
TrainFinal=Traindrop[Corr_List+['SalePrice']].dropna()
```

```python
y = TrainFinal['SalePrice']
X = TrainFinal.drop('SalePrice', axis=1)
```

```python
lrm = linear_model.LinearRegression()
lrm.fit(X, y)
```

```python
print('Değişkenler: \n', lrm.coef_)
print('Sabit değer (bias): \n', lrm.intercept_)
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)
results = sm.OLS(y, X).fit()
results.summary()
```
